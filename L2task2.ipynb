```python
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv("D:/restaurant_dataset.csv")
print(data.head(1))
```

# Task2: Cuisine Combination

## Identify the most common combinations of cuisines in the dataset.

```python
common_combinations= data.groupby('Cuisines')['Aggregate rating'].mean().sort_values(ascending=False)
top5=common_combinations.head(5)
print("Top 10 most common combinations are : ",top5)

```

```python
plt.figure(figsize=(6, 5))
top5.plot(kind='barh', color='skyblue')
plt.title("Top 5 Cuisine Combinations by Mean Aggregate Rating")
plt.xlabel("Mean Aggregate Rating")
plt.ylabel("Cuisine Combinations")
plt.show()
```

## Determine if certain cuisine combinations tend to have higher ratings.

```python
max_rating = common_combinations.iloc[0]
print("The Max Rating is : ",max_rating)
```

```python
max_rated_rest = data.loc[data['Aggregate rating'] == max_rating]
print("Restaurants having the maximum rating : ", max_rated_rest['Restaurant Name'])
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
data = pd.read_csv("D:/restaurant_dataset.csv")
print(data.head(1))
```

# Task 2: Votes Analysis
## Identify the restaurants with the highest and lowest number of votes.

```python
max_votes = data['Votes'].max()
restaurant_with_max_votes = data.loc[data['Votes'] == max_votes, 'Restaurant Name']
print("Restaurant(s) with the Maximum votes:")
print(restaurant_with_max_votes)
min_votes = data['Votes'].min()
restaurant_with_min_votes = data.loc[data['Votes'] == min_votes, 'Restaurant Name']
 
print("Restaurant(s) with the Minimum votes:")
print(restaurant_with_min_votes)
```

## Analyze if there is a correlation between the number of votes and the rating of a restaurant.

```python
correlation = data['Votes'].corr(data['Aggregate rating'])
print(f"Correlation between Votes and Aggregate Rating: {correlation:.2f}")

```

```python
## Visualize the relationship using scatterplot.
plt.figure(figsize=(10, 6))
sns.scatterplot(x=data['Votes'], y=data['Aggregate rating'], alpha=0.7, edgecolor='k')
plt.title("Relationship Between Votes and Aggregate Rating")
plt.xlabel("Number of Votes")
plt.ylabel("Aggregate Rating")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv("D:/restaurant_dataset.csv")
print(data.head(1))
```

## Task3: Price Range Distribution

##Create a histogram or bar chart to
visualize the distribution of price ranges
on the restaurants.



```python
price_range_count = data['Price range'].value_counts()
price_range_percentage = (price_range_count / price_range_count.sum()) * 100

# plot the Barchart for price range distribution.
plt.figure(figsize=(6, 4))
price_range_count.sort_index().plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Price Range Distribution', fontsize=16)
plt.xlabel('Price Range', fontsize=12)
plt.ylabel('Number of Restaurants', fontsize=12)
plt.show()
```

## Calculate the percentage of restaurants each price range category.


```python
print("Percentage of restaurants in each price range:",price_range_count)
print(price_range_percentage.sort_index())
```

```python
plt.figure(figsize=(6, 4))
plt.pie(price_range_percentage,autopct='%1.1f%%',colors=["orange","green","pink","gray"])
plt.title("Price range percentage ")
plt.show()
```
